```python
def predict_credit_risk(input_df):

    prediction = model.predict(input_df)

    probability = model.predict_proba(input_df)

    return prediction, probability
```

```python
sample = X.iloc[[0]]

prediction, probability = predict_credit_risk(sample)

print("Prediction :", prediction[0])

print("Probability :", probability[0][1])
```

```python
if prediction[0] == 1:
    print("⚠️ High Credit Risk")
else:
    print("✅ Low Credit Risk")

print(f"Risk Probability : {probability[0][1]:.2%}")
```
